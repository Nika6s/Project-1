3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience(arg):
    if arg is np.nan or arg == 'Не указано':
        return None
    year_words=['год', 'года', 'лет']
    month_words=['месяц', 'месяца', 'месяцев']
    arg_splitted = arg.split(' ')[:6]
    years = 0
    months = 0
    for i, item in enumerate(arg_splitted):
        if item in year_words:
            years = int(arg_splitted[i-1])
        if item in month_words:
            months = int(arg_splitted[i-1])
    return int(years*12 + months)

hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_experience)
work_experience = hh_data['Опыт работы (месяц)'].median()
display(f"Медианный опыт работы (в месяцах): {work_experience}")
```

```python
hh_data = hh_data.drop(['Опыт работы'], axis=1)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
# Функция для классификации городов
def classify_city(town):
    city_part = town.split(',')[0].strip()
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    if 'Москва' in city_part:
        return 'Москва'
    elif 'Санкт-Петербург' in city_part:
        return 'Санкт-Петербург'
    elif any(city in city_part for city in million_cities):
        return 'город-миллионник'
    else:
        return 'другие'

hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(classify_city)

#Функция для определения готовности к переезу
def check_relocation(reloc):
    if 'не готов к переезду' in reloc or 'не готова к переезду' in reloc:
        return False
    else:
        return True

hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(check_relocation)

# Функция для определения готовности к командировкам
def is_ready_for_business_trips(trip):
    # Проверяем, содержат ли строки ключевые слова, указывающие на готовность к командировкам
    if ('командировка' in trip):
        if ('не готов к командировкам' in trip) or('не готова к командировкам' in trip):
            return False
        else: 
            return True
    else:
        return False

hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(is_ready_for_business_trips)

# Количество соискателей из Санкт-Петербурга
applicant = hh_data[(hh_data['Город'] == 'Санкт-Петербург')]
perc_SP = round((applicant.shape[0] / hh_data.shape[0] * 100), 0)
display(f'Количество соискателей из Санкт-Петербурга {perc_SP}%')

# Количество соискателей готовых одновременно и к переездам, и к командировкам
ready_reloc_trip = hh_data[(hh_data['Готовность к переезду'] == True) & (hh_data['Готовность к командировкам'] == True)]
perc_ready = round((ready_reloc_trip.shape[0] / hh_data.shape[0]) * 100, 0)
display(f"Процент соискателей, готовых к переездам и командировкам: {perc_ready}%")
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
# Определяем типы занятости
categories = ['волонтерство', 'полная занятость', 'проектная работа', 'стажировка', 'частичная занятость']

# Добавляем новые колонки с типами занятости
for cat in categories:
    hh_data[cat] = hh_data["Занятость"].apply(lambda x: cat in x)

# Определяем виды графиков работы
categories = ['вахтовый метод', 'гибкий график', 'полный день', 'сменный график', 'удаленная работа']

# Добавляем новые колонки с видами графиков работы
for cat in categories:
    hh_data[cat] = hh_data["График"].apply(lambda x: cat in x)

# Количество людей которые ищут проектную работу и волонтёрство
mean_count_pr_rab_vol = hh_data[(hh_data['проектная работа'] == True) & (hh_data['волонтерство'] == True)].shape[0]
display(f"Количество людей которые ищут проектную работу и волонтёрство: {mean_count_pr_rab_vol}")
# Количество людей которые хотят работать вахтовым методом и с гибким графиком
mean_count_vah_gib_gr = hh_data[(hh_data['вахтовый метод'] == True) & (hh_data['гибкий график'] == True)].shape[0]
display(f"Количество людей которые хотят работать вахтовым методом и с гибким графиком: {mean_count_vah_gib_gr}")
```

Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
exchange_rate_data = pd.read_csv('data/ExchangeRates.csv', sep = ',')
exchange_rate_data = exchange_rate_data.drop(['per', 'time', 'vol'], axis=1)

# Преобразуем даты в формат datetime
exchange_rate_data['date'] = pd.to_datetime(exchange_rate_data['date'], format='%d/%m/%y')
hh_data['date'] = pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True).dt.date
hh_data['date'] = pd.to_datetime(hh_data['date'])

# Выделяем из ЗП сумму желаемой заработной платы и наименование валюты
hh_data['Зарплата']= hh_data['ЗП'].apply(lambda x: str(x).split()[0])
hh_data['Зарплата'] = hh_data['Зарплата'].astype('int64')
hh_data['currency']= hh_data['ЗП'].apply(lambda x: str(x).split()[1])

# Наименование валюты переводим в стандарт ISO
hh_data['currency'] = hh_data['currency'].replace({
    'руб.': 'RUB',
    'руб': 'RUB',
    'грн.':  'UAH',
    'USD':  'USD',
    'EUR':  'EUR',
    'бел.руб.':  'BYN',
    'KGS':  'KGS',
    'сум':  'UZS',
    'AZN':  'AZN',
    'KZT':  'KZT'
})
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True).dt.date

# Присоединяем к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты
hh_data_merge = pd.merge(hh_data, exchange_rate_data, on=['date', 'currency'], how='left')

# Заполняем признаки close и proportion значением 1 (в этих строках ЗП изначально в рублях)
hh_data_merge['close'] = hh_data_merge['close'].fillna(1)
hh_data_merge['proportion'] = hh_data_merge['proportion'].fillna(1)
# Рассчитываем желаемую медианную заработную плату соискателей
hh_data_merge['ЗП (руб)'] = hh_data_merge['Зарплата'] * hh_data_merge['close']/hh_data_merge['proportion']
median_pay = round((hh_data_merge['ЗП (руб)'].median())/1000, 0)
display(f'Желаемая медианная заработная плата соискателей: {median_pay} тыс. руб.')

```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# ваш код здесь
# Создаем фигуру и оси
plt.figure(figsize=(10, 6))

# Гистограмма для признака 'Возраст' в логарифмическом масштабе
log_mkad_km= np.log(hh_data_merge_duplicates_1['Возраст'] + 1)
histplot = sns.histplot(log_mkad_km , bins=30,)
histplot.set_xlabel('Логарифм возраста')
histplot.set_ylabel('Частота')
histplot.set_title('Распределение логарифма возраста');

# Логарифмирование возраста
log_age = np.log(hh_data_merge_duplicates_1['Возраст'])
mean_log_age = log_age.mean()
std_log_age = log_age.std()

histplot.axvline(mean_log_age, color='k', lw=2, label='Mean')
histplot.axvline(mean_log_age + 3 * std_log_age, color='r', lw=2, linestyle='--', label='Mean + 3σ')
histplot.axvline(mean_log_age - 3 * std_log_age, color='r', lw=2, linestyle='--', label='Mean - 3σ')
histplot.axvline(mean_log_age + 4 * std_log_age, color='g', lw=2, linestyle='-.', label='Mean + 4σ')
plt.legend()

# Показываем график
plt.show()


# Определение и удаление выбросов
z_scores = (log_age - mean_log_age) / std_log_age
outliers = hh_data_merge_duplicates_1[(z_scores > 4) | (z_scores < -3)]
df_cleaned = hh_data_merge_duplicates_1[(z_scores <= 4) & (z_scores >= -3)]

# Вывод таблицы с выбросами и оценка возраста соискателей, попадающих под категорию выбросов
display("Выбросы:")
display(outliers['Возраст'])

# Функция для поиска выбросов методом z-отклонений
def outliers_z_score_mod(data, feature, left=3, right=4, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# Применяем функцию
outliers, cleaned = outliers_z_score_mod(hh_data_merge_duplicates_1, 'Возраст', log_scale=True)
display(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
display(f'Результирующее число записей: {cleaned.shape[0]}')
```

Логарифмическое распределение асимметрично в правую сторону(положительная асимметрия). Это говорит о том, что в данных присутствуют редкие, но значительно большие значения по сравнению с остальными, и это имеет важные последствия для анализа и интерпретации этих данных.
